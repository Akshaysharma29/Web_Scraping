```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup as BS
import requests
```

```python
url = "https://www.imdb.com/chart/top?ref_=nv_mv_250_6"
```

```python
r = requests.get(url)
print(type(r))
```

```python
soup = BS(r.text,'html.parser')
```

```python
# print(soup)
```

```python
print(type(soup))
```

```python
TITLES = []
RATINGS = []
URLS = []
YEARS = []

DIRECTOR = []
ACTORS = []
```

```python
divs = soup.find_all('div', {'class' : 'lister'})
print(len(divs))
```

```python
div = divs[0]
rows = div.find_all('tr')
print(type(rows))
print(len(rows))
```

```python
# print(rows)
```

```python
print(rows[0])
```

```python
print(rows[1])
```

```python
title_column= rows[1].find('td',{'class':'titleColumn'})
print(title_column)
print('----------')
anchor = title_column.find('a')
print(anchor)
print('----------')
print(anchor.text)
print('----------')
print(anchor['href'])
print('----------')
print(anchor['title'].split(',')[0].split('(')[0])
print('----------')
print(','.join(anchor['title'].split(',')[1:]))
print('----------')
print('http://www.imdb.com' + anchor['href'])
print('----------')
print(title_column.find('span').text.split('(')[1].split(')')[0])
print('-------------------------')
rate_column= rows[1].find('td',{'class':'ratingColumn imdbRating'})
print(rate_column.find('strong').text)
```

```python
count=0

for row in rows[1:]:
    count+=1
    title_Column = row.find('td', {'class':'titleColumn'})
    anchor = title_Column.find('a')
    movie_title = anchor.text
    movie_rel_url = anchor['href']
    director = anchor['title'].split(',')[0].split('(')[0]
    actors = ','.join(anchor['title'].split(',')[1:])
    movie_url = 'http://www.imdb.com' + movie_rel_url
    year_data = title_Column.find('span')
    year = year_data.text.split('(')[-1].split(')')[0]
    
    rating_Column = row.find('strong')
    rating = rating_Column.text
    #print movie_title, year, rating, movie_url
    #print '--------------'
#     TITLES.append(str(movie_title.encode('utf-8', 'ignore')))
#     URLS.append(str(movie_url.encode('utf-8', 'ignore')))
#     RATINGS.append(str(rating.encode('utf-8', 'ignore')))
#     YEARS.append(year)
    
#     ACTORS.append(str(actors.encode('utf-8', 'ignore')))
#     DIRECTOR.append(str(director.encode('utf-8', 'ignore')))
    TITLES.append(str(movie_title))
    URLS.append(str(movie_url))
    RATINGS.append(str(rating))
    YEARS.append(year)
    
    ACTORS.append(str(actors))
    DIRECTOR.append(str(director))
```

```python
print(len(TITLES), len(URLS), len(RATINGS), len(YEARS), len(ACTORS), len(DIRECTOR))
```

```python
init_df = pd.DataFrame({'Title': TITLES, 'URL': URLS, 'Rating': RATINGS, 'Year': YEARS, 'Actors': ACTORS, 'Director': DIRECTOR})
print(type(init_df))
```
